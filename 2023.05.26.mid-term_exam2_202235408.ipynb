```python
import pandas as pd
import numpy as np
```

```python
mushroom_data = pd.read_csv("C:/ai_dataset/mushrooms.csv")
```

```python
mushroom_data.drop(columns=['class'], inplace=True)
```

```python
mushroom_data
```

```python
mushroom_data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for i in mushroom_data.columns :
    mushroom_data[i] = le.fit_transform(mushroom_data[i])
```

```python
mushroom_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(data = mushroom_data, x="gill-size")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Label")
```

```python
from sklearn.model_selection import train_test_split

X = mushroom_data.drop('gill-size',axis=1)
y = mushroom_data['gill-size']

X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, shuffle=True, random_state=12)
```

```python
import numpy as np

mean = np.mean(X_train, axis = 0)

std = np.std(X_train, axis = 0)

X_train = (X_train - mean) / std
X_test = (X_test - mean) / std
```

```python
print(np.any(np.isnan(X_train)))
print(np.any(np.isnan(X_test)))
print(np.any(np.isinf(X_train)))
print(np.any(np.isinf(X_test)))
```

```python
X_train = pd.DataFrame(X_train).fillna(0).values
X_test = pd.DataFrame(X_test).fillna(0).values
```

```python
print(np.any(np.isnan(X_train)))
print(np.any(np.isnan(X_test)))
print(np.any(np.isinf(X_train)))
print(np.any(np.isinf(X_test)))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(40, activation = 'relu', input_shape = (21,)))
model.add(Dense(20, activation = 'relu'))
model.add(Dense(2, activation = 'sigmoid')) 

model.compile(optimizer = 'rmsprop', loss='mse', metrics = ['mae'])

model.summary()
```

```python
N_EPOCHS = 10

history = model.fit(X_train, y_train, epochs = N_EPOCHS, batch_size=64, validation_split=0.2)
```

```python
model.evaluate(X_test, y_test)
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(N_EPOCHS), history.history['loss'], label='train loss')
plt.plot(range(N_EPOCHS), history.history['val_loss'], label='test loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(N_EPOCHS), history.history['mae'], label='mae')
plt.plot(range(N_EPOCHS), history.history['val_mae'], label='test mae')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
