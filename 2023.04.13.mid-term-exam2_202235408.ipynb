<h1>중간고사 실기 시험 2</h1>

<h2>Iris Dataset</h2>

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

iris = load_iris()
df_iris_feature = iris.feature_names
df_iris_data = pd.DataFrame(iris.data, columns = df_iris_feature)
df_iris_data['species'] = iris.target
```

```python
df_iris_data['species'] = df_iris_data['species'].map({0:"setosa", 1:"versicolor", 2:"virginica"})

df_iris_data.head()
```

<h2>Regression</h2>

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
drop_column = ['species', 'petal width (cm)']

X = df_iris_data.drop(df_iris_data[drop_column], axis=1)
```

```python
y = df_iris_data['petal width (cm)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

<h3>SVM</h3>

```python
from sklearn.svm import SVR

model = SVR()
model.fit(X_train, y_train)

sy_preds = model.predict(X_test)
mse = np.sqrt(mean_squared_error(sy_preds, y_test))
print('mse : ', mse)
```

```python
plt.scatter(y_test, sy_preds)
plt.xlabel("True Petal Width (cm)")
plt.ylabel("Predicted Petal Width (cm)")
plt.show()
```

<h3>LR</h3>

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

ly_preds = model.predict(X_test)
mse = np.sqrt(mean_squared_error(ly_preds, y_test))
print('mse : ', mse)
```

```python
plt.scatter(y_test, ly_preds)
plt.xlabel("True Petal Width (cm)")
plt.ylabel("Predicted Petal Width (cm)")
plt.show()
```

<h3>DT</h3>

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(X_train, y_train)

dy_preds = model.predict(X_test)
mse = np.sqrt(mean_squared_error(dy_preds, y_test))
print('mse : ', mse)
```

```python
plt.scatter(y_test, dy_preds)
plt.xlabel("True Petal Width (cm)")
plt.ylabel("Predicted Petal Width (cm)")
plt.show()
```

<h3>RF</h3>

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)

ry_preds = model.predict(X_test)
mse = np.sqrt(mean_squared_error(ry_preds, y_test))
print('mse : ', mse)
```

```python
plt.scatter(y_test, ry_preds)
plt.xlabel("True Petal Width (cm)")
plt.ylabel("Predicted Petal Width (cm)")
plt.show()
```
